```python
from googleapiclient.discovery import build
import pandas as pd

# Replace with your own API key
api_key = "AIzaSyDxa1eK21TK4kdvp3q7aA-_ENPg-Gk2jdU"
youtube = build('youtube', 'v3', developerKey=api_key)

def get_uploads_playlist_id(channel_id):
    request = youtube.channels().list(
        part="contentDetails",
        id=channel_id
    )
    response = request.execute()
    return response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

def get_videos_from_playlist(playlist_id, max_videos=100):
    videos = []
    next_page_token = None
    while len(videos) < max_videos:
        request = youtube.playlistItems().list(
            part="snippet",
            playlistId=playlist_id,
            maxResults=min(50, max_videos - len(videos)),
            pageToken=next_page_token
        )
        response = request.execute()
        videos += response['items']
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break
    return videos

def get_video_stats(video_ids):
    stats = []
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="statistics,snippet",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute()
        stats += response['items']
    return stats
```

```python
# Example Channel (MrBeast)
channel_id = "UCX6OQ3DkcsbYNE6H8uQQuVA"

uploads_playlist_id = get_uploads_playlist_id(channel_id)
videos = get_videos_from_playlist(uploads_playlist_id, max_videos=100)

video_ids = [video['snippet']['resourceId']['videoId'] for video in videos]
video_stats = get_video_stats(video_ids)

# Process data
data = []
for video in video_stats:
    stats = video['statistics']
    snippet = video['snippet']
    data.append({
        'title': snippet['title'],
        'views': int(stats.get('viewCount', 0)),
        'likes': int(stats.get('likeCount', 0)),
        'comments': int(stats.get('commentCount', 0)),
        'title_length_chars': len(snippet['title']),
        'title_length_words': len(snippet['title'].split())
    })

df = pd.DataFrame(data)
df.to_csv("youtube_video_data.csv", index=False)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: title length vs views
plt.figure(figsize=(10,6))
sns.scatterplot(x='title_length_chars', y='views', data=df)
plt.title("Title Length (Characters) vs Views")
plt.xlabel("Title Length (Characters)")
plt.ylabel("View Count")
plt.grid(True)
plt.show()

# Correlation
correlation = df['title_length_chars'].corr(df['views'])
print(f"Correlation between title length and views: {correlation:.3f}")
```

```python
import pandas as pd

# Load the channel IDs from the CSV file
df_channels = pd.read_csv('collected_channel_ids.csv')

# Check the first few entries to confirm
print(df_channels.head())
```

```python
def fetch_channel_video_data(channel_id, max_videos=50):
    try:
        # Get the uploads playlist of the channel
        playlist_id = get_uploads_playlist_id(channel_id)
        # Fetch videos from the playlist
        videos = get_videos_from_playlist(playlist_id, max_videos=max_videos)
        video_ids = [v['snippet']['resourceId']['videoId'] for v in videos]
        video_stats = get_video_stats(video_ids)

        # Collect video data
        data = []
        for video in video_stats:
            stats = video['statistics']
            snippet = video['snippet']
            data.append({
                'channel_id': channel_id,
                'title': snippet['title'],
                'views': int(stats.get('viewCount', 0)),
                'likes': int(stats.get('likeCount', 0)),
                'comments': int(stats.get('commentCount', 0)),
                'title_length_chars': len(snippet['title']),
                'title_length_words': len(snippet['title'].split())
            })
        return data
    except Exception as e:
        print(f"Error processing channel {channel_id}: {e}")
        return []
```

```python
def get_uploads_playlist_id(channel_id):
    try:
        request = youtube.channels().list(
            part="contentDetails",
            id=channel_id
        )
        response = request.execute()
        uploads_playlist_id = response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
        return uploads_playlist_id
    except Exception as e:
        print(f"Error getting uploads playlist for channel {channel_id}: {e}")
        return None
```

```python
def get_video_stats(video_ids):
    try:
        request = youtube.videos().list(
            part="snippet,statistics",
            id=','.join(video_ids)
        )
        response = request.execute()
        return response['items']
    except Exception as e:
        print(f"Error fetching video stats: {e}")
        return []
```

```python
all_data = []

for i, channel_id in enumerate(df_channels['channel_id']):
    print(f"Fetching data for channel {i + 1}/{len(df_channels)}: {channel_id}")
    channel_data = fetch_channel_video_data(channel_id, max_videos=30)  # Adjust `max_videos` as needed
    all_data.extend(channel_data)

# Convert the collected data to a pandas DataFrame
df_video_data = pd.DataFrame(all_data)

# Show the first few rows of the data
df_video_data.head()
```
